Notebook for testing and demonstrating the baseline Shor's algorithm implementation.

```python
from quantumthreattracker.algorithms.baseline_shor import (
    BaselineShor,
    BaselineShorParameters,
)

params = BaselineShorParameters(protocol="RSA", key_size=4)
shor = BaselineShor(params)
circuit = shor.create_qiskit_circuits()[0][0]
circuit.draw("latex")
```

```python
from qsharp.estimator import EstimatorParams
from qsharp.interop.qiskit import estimate
from qsharp_widgets import EstimateDetails

params = EstimatorParams()
result = estimate(circuit, params)

EstimateDetails(result)
```
