```python
from qsharp.estimator import EstimatorParams

from quantumthreattracker.algorithms import CryptParams
from quantumthreattracker.algorithms.rsa.chevignard import (
    Chevignard,
    ChevignardParams,
)
```

```python
crypt_params = CryptParams(protocol="RSA", key_size=2048)

algorithm = Chevignard(crypt_params, ChevignardParams())
estimator_params = EstimatorParams()
estimator_params.qubit_params.name = "qubit_gate_ns_e3"

algorithm.estimate_resources_azure(estimator_params)
```
